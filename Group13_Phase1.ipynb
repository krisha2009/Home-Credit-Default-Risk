# Project Abstract
<p>
Home Credit Group is in pursuit of solutions to service individuals with little to no credit history in the loan industry. Home Credit Group looks to use alternative data that will help them decide if they are able to lend to these individuals. With the utilization of machine learning models, Home Credit Group hopes to be able to serve unbanked people with confidence. Our team aims to successfully aid in this process by utilizing our knowledge of machine learning pipelines, machine learning models, and data analysis. Specifically, we intend to deploy, train, and optimize linear regression models, random forest models, and a neural network using adagrad. We will measure our success, first, by the metric Home Credit Group uses Area Under Curve. We will also use accuracy score, and f1 score, since this problem is rooted in a classification problem and false positives and negatives are essential. We will also employ L1, L2, and MSE loss functions to help guide our model to an optimized implementation. 
</p>

# Machine Learing Algorithms

## Machine Learning - Algorithm Selection
Given the Home Credit Groupâ€™s goal of being able to validate the lendability of an individual with secondary credit data, </br>we must employ appropriate functions that accuratly predict classifying an individual as a lendable or un-lendable individual. </br>That being said, using machine learning algorithms such as linear regression, random forest, and trained neural networks using the </br> optimization function of adagrad seem logical in our goal of classification and minimizing AUC differential from the target.

### Logistic Regression

Logistic Regression seems to be a logical option for this data set. Since logistic regression has great flexibility in data sets with relatively lower number of features,</br> we will be able to explore the data and test data analysis methods with this function and get a better picture of the problem itself. </br>Additionally, this method will force us to refine our methods of handling categorical and numerical data since it has no immunity </br>to these characteristic of the data.

### Random Forest
Random Forest seems to be beneficial in this scenario, because there is both categorical and numerical data. From our findings in class we understand that Random Forest handles this data well and can find the intersection between numerical and categorical data easily. Also we believe that this function will allow us to explore the data in meaningful ways. Allowing us to take these lessons into other, more preformant models like a neural network.

### Neural Network (Adagrad)

After previous data analysis and hyper parameter tuning it seems that the most accurate results for this data set will yield from a neural network. We specifically believe that the adagrad optimization will be useful because of its ability to to predict validation functions like area under the curve which is the main scoring mechanism deployed. We believe that adagrad will be a good solution since it is based on gradient descent and we are aiming for a specific value from this descent.

# Machine Learning - Metrics

## Area Under The Curve (AUC)

$\text{AUC} = \int_{-\infty}^{\infty} \text{TPR}(f) , \text{d}\text{FPR}(f)$

## F1 Score

$F_1 = 2 \cdot \frac{precision \cdot recall}{precision + recall}$

## Accuracy Score

$accuracy = \frac{TP + TN}{TP + TN + FP + FN}$
